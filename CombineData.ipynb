```python
import pandas as pd
import datetime

pd.set_option('display.max_columns', 85)
```

```python
redfin_df_orig = pd.read_csv('processed/redfin_month_rollup_2023-06-08.tsv', sep='\t')
redfin_df_orig
```

```python
redfin_df = redfin_df_orig[redfin_df_orig['months_averaged_redfin'].isin([1, 3, 6, 12, 18])]
redfin_df
```

```python
prcyd_df = pd.read_csv('processed/prycd_county_data_2023-06.tsv', sep='\t')
prcyd_df
```

```python
merged_df = redfin_df.merge(prcyd_df, how='inner', on='county_qualified_name')
merged_df
```

```python
combined_df = merged_df.dropna(subset=['property_count', 'inventory_redfin']).copy()
combined_df
```

```python
combined_df['county_unique'] = combined_df['county_qualified_name'] + ' - ' + combined_df['months_averaged_redfin'].astype(str)
cols = ['county_unique'] + [col for col in combined_df.columns if col != 'county_unique']

combined_df = combined_df[cols]
combined_df
```

```python
export_df = combined_df.drop(columns=['County', 'State', 'state_abbrev', 'county_qualified_name'])
```

```python
today = datetime.date.today()
date_string = today.strftime("%Y-%m-%d")
file_path_tsv = f"processed/combined_county_data_{date_string}.tsv"
file_path_csv = f"processed/combined_county_data_{date_string}.csv"
```

```python
export_df.to_csv(file_path_tsv, index=False, sep ='\t')
export_df.to_csv(file_path_csv, index=False)
```
