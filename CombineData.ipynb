```python
import pandas as pd
import datetime

pd.set_option('display.max_columns', 85)
```

```python
redfin_df_orig = pd.read_excel('processed/redfin_month_rollup_2023-07-05.xlsx')
redfin_df_orig
```

```python
# redfin_df = redfin_df_orig[redfin_df_orig['months_averaged_redfin'].isin([1, 3, 6, 12, 18])]
redfin_df = redfin_df_orig[redfin_df_orig['months_averaged_redfin'].isin([3])]
redfin_df
```

```python
prcyd_df = pd.read_excel('processed/prycd_county_data_2023-07.xlsx')
prcyd_df
```

```python
merged_df = redfin_df.merge(prcyd_df, how='outer', on='county_qualified_name')
merged_df
```

```python
today = datetime.date.today()
date_string = today.strftime("%Y-%m-%d")
```

```python
combined_df = merged_df
combined_df['county_unique'] = combined_df['county_qualified_name']
combined_df = combined_df.assign(generated_date = date_string)
cols = ['county_unique'] + [col for col in combined_df.columns if col != 'county_unique']

combined_df = combined_df[cols]
combined_df
```

```python
combined_df.drop_duplicates(inplace=True)
```

```python
file_path = f"processed/combined_county_data_{date_string}.xlsx"
```

```python
combined_df.to_excel(file_path, index=False)
```
