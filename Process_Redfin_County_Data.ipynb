```python
# import csv which is tab separated

df_original = pd.read_csv('/content/drive/My Drive/Land/county_market_tracker.csv', sep='\t')
```

```python
df_original.info()
```

```python
df_copy = df_original.copy()
```

```python
df_copy
```

```python
df = df_copy.loc[:, ['period_begin', 'region', 'state', 'state_code', 'property_type', 'homes_sold', 'new_listings', 'inventory', 'median_dom', 'months_of_supply', 'avg_sale_to_list', 'last_updated']]
```

```python
df = df.sort_values(by='period_begin', ascending=False)
df = df.reset_index(drop=True)
df = df.rename(columns={'region':'county', 'period_begin': 'month'})
df
```

```python
# Convert the 'month' column to a datetime object
df['month'] = pd.to_datetime(df['month'])

filtered_df = df[df['month'] >= '2021-01-01']
df_residential = filtered_df[df['property_type'] == 'All Residential']

df_residential
```

```python
df_2 = df_residential.copy()
```

```python
df = df_2

df['county_name'] = df['county'].apply(lambda x: x.split(',')[0] if ',' in x else x)
```

```python
df
```

```python
df.to_csv('redfin_processed_output.csv', encoding = 'utf-8-sig') 
```
