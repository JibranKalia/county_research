```python
import pandas as pd
import datetime

pd.set_option('display.max_columns', 85)
```

```python

df_original = pd.read_csv('data/redin_county_data_2023_05_14.tsv', sep='\t')
```

```python
df_copy = df_original.copy()
```

```python
df_copy
```

```python
df = df_copy.loc[:, ['period_begin', 'region', 'state', 'state_code', 'property_type', 'homes_sold', 'new_listings', 'inventory', 'median_dom', 'months_of_supply', 'avg_sale_to_list', 'last_updated']]
```

```python
df = df.sort_values(by='period_begin', ascending=False)
df = df.reset_index(drop=True)
df = df.rename(columns={'region':'county', 'period_begin': 'month'})
df
```

```python
# Convert the 'month' column to a datetime object
df['month'] = pd.to_datetime(df['month'])
df['county_name'] = df['county'].apply(lambda x: x.split(',')[0] if ',' in x else x)

filtered_df = df[df['month'] >= '2022-01-01']
df_residential = filtered_df[filtered_df['property_type'] == 'All Residential']

df_residential
```

```python
df = df_residential.copy()

# Drop the 'county' and 'state' columns
df = df.drop(columns=['county', 'state'])

# Rearrange the columns
df = df[['month', 'county_name', 'state_code', 'property_type', 'homes_sold', 'new_listings', 'inventory', 
         'median_dom', 'months_of_supply', 'avg_sale_to_list', 'last_updated']]

df
```

```python
# Get today's date
today = datetime.date.today()
date_string = today.strftime("%Y-%m-%d")
file_path = f"data/redfin_processed_output_{date_string}.csv"
df.to_csv(file_path, index=False)
```
