```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

pd.set_option('display.max_columns', 85)
```

```python
# import csv which is tab separated

df_original = pd.read_csv('/content/drive/My Drive/Land/county_market_tracker.csv', sep='\t')
```

```python
df_original.info()
```

```python
df_copy = df_original.copy()
```

```python
df = df_copy.loc[:, ['period_begin', 'region', 'state', 'state_code', 'homes_sold', 'new_listings', 'inventory', 'median_dom', 'months_of_supply', 'avg_sale_to_list']]
```

```python
df = df.sort_values(by='period_begin', ascending=False)
df = df.reset_index(drop=True)
df = df.rename(columns={'region':'county', 'period_begin': 'month'})
df
```

```python
# Convert the 'month' column to a datetime object
df['month'] = pd.to_datetime(df['month'])

filtered_df = df[df['month'] >= '2022-01-01']
filtered_df
```
