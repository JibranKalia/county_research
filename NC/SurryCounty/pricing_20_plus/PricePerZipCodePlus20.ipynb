```python
import pandas as pd
df = pd.read_csv('properties_with_zipcode.csv')
df
```

```python
# Group by 'zipcode' and calculate mean of 'price_per_acre' and count of data points
grouped_df = df.groupby('zip').agg(
    average_price_per_acre=pd.NamedAgg(column='price_per_acre', aggfunc='mean'),
    count=pd.NamedAgg(column='price_per_acre', aggfunc='count')
)

grouped_df
```

```python
grouped_df.to_excel('price_per_zipcode.xlsx')
```
